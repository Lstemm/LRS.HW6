b) Use the FacetGrid from seaborn to create line plots to represent house_price_index for each state. Please make the line color red, set ticks on x axis for years ['1980', '2000', '2020'], which you may have to convert date type again, put those years as labels, and set the titles for each small plot to their state names. In addition, set the y label to 'house price index' and remove the x label since it's obvious that these are years.

```python
g = sns.FacetGrid(states, col="state", col_wrap=2, height=4, aspect=1)
g.map(sns.lineplot, 'date', 'house_price_index', color='red')

# Set x-axis ticks and labels
years_to_label = ['1980', '2000', '2020']
datetime_ticks = [pd.to_datetime(f'{year}-01-01') for year in years_to_label]

for ax in g.axes.flat:
    ax.set_xticks(datetime_ticks)
    ax.set_xticklabels(years_to_label)
    ax.set_xlabel('') # Remove x label as requested
    ax.set_ylabel('house price index') # Set y label as requested
    ax.set_title(ax.get_title().split('=')[1].strip()) # Set title to state name
```

```python
#Your code goes here.
world_cup = pd.read_csv('/content/drive/MyDrive/DATA/WorldCupMatches.csv')
world_cup.head()
```

Make sure that you only write four columns ('Home Team Initials', 'Away Team Initials', 'weight', 'HomeGoalTotal') from the data frame, team_pairs, to a csv file

```python
#your code goes here. Just leaving an example output of an intermediate step for you to check the last two columns
team_pairs = world_cup.groupby(['Home Team Initials', 'Away Team Initials']).count().reset_index()
team_pairs.head()
```

```python
#Part a
all_teams = pd.concat([world_cup['Home Team Initials'], world_cup['Away Team Initials']])
weight = all_teams.value_counts()
weight.head()
```

```python
#Part b
home_goal_total = world_cup.groupby('Home Team Initials')['Home Team Goals'].sum().reset_index()
home_goal_total = home_goal_total.rename(columns={'Home Team Goals': 'HomeGoalTotal'})
display(home_goal_total.head())
```

```python
#Part c
team_pairs = world_cup.groupby(['Home Team Initials', 'Away Team Initials']).size().reset_index(name='weight')
team_pairs = pd.merge(team_pairs, home_goal_total, on='Home Team Initials', how='left')
team_pairs = team_pairs[['Home Team Initials', 'Away Team Initials', 'weight', 'HomeGoalTotal']]
display(team_pairs.head())
```

```python
#Part d
team_pairs.to_csv('team_pairs.csv', index=False)
```
